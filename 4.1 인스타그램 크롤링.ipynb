### 게시글 정보 가져오기

```python
import re
from bs4 import BeautifulSoup
def get_content(driver):
    #1.현재페이지 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    #2.본문내용 가져오기
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content=''
    
    #3.본문내용에서 해시태그 가져오기(정규식활용)
    tags = re.findall(r'#[^\s#,\\]+', content) #   r'#(샵으로시작)[^(제외)\s(빈공간)#(샵),(콤마)\\(역슬래쉬)]+',
    
    #4.작성일자정보 가져오기
    date = soup.select('time.FH9sR.Nzb55')[0]['datetime'][:10] #[:10] => datetime속성값에서 연-월-일 정보만 활용
    
    #5.좋아요수 가져오기
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1]
    except:
        like = 0
    
    #6.위치정보 가져오기
    try:
        place = soup.select('div.JF9hh > a')[0].text
    except:
        place=''
    
    #7.수집한 정보 저장하기
    data = [content, date, like, place, tags]
    
    return data
```

### 다음 게시글 열기

```python
def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(3)
```

### 여러 게시글 정보 수집하기

```python
result=[]
```

```python
target=100
```

```python
for i in range(target):
    try:
        data = get_content(driver) #게시글 정보 가져오기
        print(data)
        result.append(data)
        move_next(driver) #다음게시물로 이동
    except:
        time.sleep(2)
        move_next(driver)
print(result)
```
