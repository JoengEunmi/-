### 워드클라우드 만드는 과정

```python
import pandas as pd
raw_total = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/crawling_jejudoMatJip.xlsx')
raw_total.tags
```

```python
#해시태그를 통합저장
tags_total =[]

for tags in raw_total['tags']:
    tags_list = tags[2:-2].split("', '")  # tags[2:-2] : 맨처음 [' 와 맨마지막 '] 빼고 ','로 split
    for tag in tags_list: 
        tags_total.append(tag)
```

```python
tags_total
```

### 해시태그 출현 빈도 집계

```python
#빈도수 집계
from collections import Counter
tags_counts = Counter(tags_total)
```

```python
#가장 많이 사용된 해시태그 확인
tags_counts.most_common()
```

### 막대차트로 해시태그 살펴보기

```python
#시각화 라이브러리 임포트 및 글꼴 설정
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import sys

if sys.platform in ['win32', 'win64']: #한글 깨짐 현상을 방지하기위해 한글 글꼴 지정
    font_name = 'malgun gothic'
elif sys.platform == 'darwin':
    font_name = 'AppleGothic'

rc('font', family = font_name)
```

```python
#데이터 준비
tag_counts_df = pd.DataFrame(tag_counts_selected.most_common(30)) #상위 30개 데이터 추출
tag_counts_df.columns = ['tags', 'counts'] #칼럼명 지정
```

```python
#막대차트 그리기
plt.figure(figsize=(10,8))
sns.barplot(x='counts', y='tags', data=tag_counts_df)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import platform

#워드 클라우드에서는 글꼴의 경로를 직접 지정해야 한다
if platform.system() == 'Windows': #윈도우의 경우
    font_path = 'C:/Windows/Fonts/malgun.ttf'
elif platform.system() == 'Darwin': #macOS의 경우
    font_path = '/Users/$USER/Library/Fonts/AppleGothic.ttf'
```

```python
#워드 클라우드 만들기
wordcloud = WordCloud(font_path = font_path, #사용하고자 하는 글꼴의 경로
                     background_color = 'white', #배경색
                     max_words = 100, #최대 몇개의 단어를 나타낼 것인지
                     relative_scaling = 0.3, #워드 클라우드 내 글자들의 상대적인 크기 
                     #(최소0~최대1, 0에 가까울수록 빈도수의 순위, 1에 가까울수록 빈도수에 더 큰 영향) 
                     width = 800, #워드클라우드 가로 이미지 크기
                     height = 400, #워드클라우드 세로 이미지 크기
                     ).generate_from_frequencies(tag_counts_selected) 
plt.figure(figsize = (15,10)) #최종 출력물의 크기
plt.imshow(wordcloud)
plt.axis('off') 
plt.savefig('C:/Users/name_/Desktop/데이터분석/files/tag_wordcloud.png') #원하는 경로에 워드 클라우드 이미지 파일 저장 가능
```
