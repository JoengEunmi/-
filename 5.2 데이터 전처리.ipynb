### 서울시 스타벅스 매장 목록, 인구, 사업체 데이터에 시군구명, 시군구 코드 추가

```python
import pandas as pd
```

```python
#서울시 스타벅스 매장 목록이 담긴 엑셀파일 불러오기
seoul_starbucks = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/5장/seoul_starbucks.xlsx', header=0)
seoul_starbucks.head()
```

```python
#스타벅스 주소 정보에서 시군구명 추출
sgg_names = []
for address in seoul_starbucks['주소']:
    sgg = address.split()[1]
    sgg_names.append(sgg)
seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()
```

```python
#엑셀로 저장
seoul_starbucks.to_excel('C:/Users/name_/Desktop/데이터분석/files/5장/seoul_starbucks_list.xlsx', index=False)
```

### 스타벅스 분석 데이터 만들기

```python
#시군구 목록데이터 불러오기
seoul_sgg = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/5장/seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
#서울시 스타벅스 매장 목록 데이터 불러오기
seoul_starbucks = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/5장/seoul_starbucks_list.xlsx')
seoul_starbucks.head()
```

```python
#시군구별 스타벅스 매장 수 세기
starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구명',
                                                 values = '매장명',
                                                 aggfunc = 'count').rename(columns = {'매장명':'스타벅스_매장수'})
starbucks_sgg_count.head()
```

```python
#서울시 시군구 목록 데이터에서 스타벅스 매장수 데이터를 병합
seoul_sgg = pd.merge(seoul_sgg, starbucks_sgg_count,
                   how='left',
                   on = '시군구명')
seoul_sgg.head()
```

```python
#서울시 시군구별 인구통계 데이터 불러오기
seoul_sgg_pop = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/5장/sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
#서울시 시군구 목록 데이터에 서울시 시군구별 인구통계 데이터 병합
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop,
                    how='left',
                    on = '시군구명')
seoul_sgg.head()
```

```python
#서울시 시군구 목록 데이터에 서울시 시군구별 사업체수 통계 데이터 병합
seoul_sgg_biz = pd.read_excel('C:/Users/name_/Desktop/데이터분석/files/5장/sgg_biz.xlsx')
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_biz, 
                    how = 'left',
                    on = '시군구명')
seoul_sgg.head()
```

```python
#병합 결과를 엑셀 파일로 저장
seoul_sgg.to_excel('C:/Users/name_/Desktop/데이터분석/files/5장/seoul_sgg_stat.xlsx', index = False)
```
